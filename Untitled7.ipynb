```python
import emoji
import contractions
import re
```

```python
# Nama
# NIM
# 19/10/2023

# Ketik Manual dari picture yang di share

# membersihkan teks dalam tweet.
def cleansing(tweet):
  # mengganti semua kemunculan string 'RT ' dengan string kosong ('') dalam teks tweet.
  t1 = re.sub('RT\s','',tweet)

  # mengganti semua kemunculan username Twitter dalam teks t1 dengan string kosong ('').
  # '\B' digunakan untuk mencocokkan tepi kata
  t2 = re.sub('\B@\w+','',t1)

  # menggantikan semua emoji dalam teks t2 dengan teks yang sesuai.
  t3 = emoji.demojize(t2)

  # mengganti semua URL yang dimulai dengan 'http' atau 'https' dalam teks t3 dengan string kosong ('').
  t4 = re.sub('(http|https):\/\/\S+','',t3)

  # menghilangkan satu atau lebih tanda pagar '#' dalam teks t4.
  t5 = re.sub('#+','',t4)

  # menyimpan teks t5 dalam huruf kecil (lowercase).
  t6= t5.lower()

  # mengganti semua pengulangan karakter yang terdiri dari dua atau lebih karakter dengan hanya dua karakter yang sama.
  t7 = re.sub(r'(.)\1+',r'\1\1', t6)

  # menghilangkan pengulangan tanda tanya, titik, atau tanda seru yang muncul berurutan dalam teks t7.
  t8 = re.sub(r'[\?\.\!]+(?=[\?.\!])','',t7)

  # digunakan untuk mengganti semua karakter yang bukan huruf alfabet (a-z dan A-Z) dengan spasi. Hasilnya akan disimpan di t9.
  t9 = re.sub(r'[^a-zA-Z]', ' ', t8)

  t10 = contractions.fix(t9)

  return t10
```

```python
for i, r in df.iterrows():
  y = cleansing(r['full_text'])
  df1.loc[i, 'full_text'] = y
```

```python
!pip install nltk

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
from nltk.corpus import stopwords

nltk.download('wordnet')
from nltk import pos_tag
from nltk.corpus import wordnet

nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
```

```python
pos_dict = {'J':wordnet.ADJ, 'v':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}
```

```python
# Nama
# NIM
# 19/10/2023

def token_stop_pos(text):
  tags = pos_tag(word_tokenize(text))
  newlist = []
  for word, tag in tags:
    if word.lower() not in set(stopwords.words('english')):
      newlist.append(tuple([word, pos_dict.get(tag[0])]))
  return newlist
```

```python
df1['POS tagged'] = df1['full_text'].apply(token_stop_pos)
```

```python
df1
```

```python
# Nama
# NIM
# 19/10/2023

from textblob import TextBlob

def getSubjectivity(tagged_words):
  text = ' '.join([word for word, pos_tag in tagged_words])
  return TextBlob(text).sentiment.subjectivity

def getPolarity(tagged_words):
  text = ' '.join([word for word, pos_tag in tagged_words])
  return TextBlob(text).sentiment.polarity

def analyze(score):
  if score<0:
    return 'Negative'
  elif score==0:
    return 'Neutral'
  else:
    return 'Positive'
```

```python
# Nama
# NIM
# 19/10/2023

final_data = pd.DataFrame(df1[['user_id_str','POS tagged']])
final_data
```

```python
# Nama
# NIM
# 19/10/2023

final_data['Subjectivity'] = final_data['POS tagged'].apply(getSubjectivity)
final_data['Polarity'] = final_data['POS tagged'].apply(getPolarity)
final_data['TextBlob'] = final_data['Polarity'].apply(analyze)
```
